```python
import warnings
import pandas as pd
import numpy as np

warnings.filterwarnings('ignore')
```

### Điền khuyết bằng mean

```python
# Tạo DataFrame mẫu với một số giá trị bị khuyết
data = {
    'A': [1, 2, 'null', 4, 5],
    'B': [np.nan, 2, 3, 4, 'unknown'],
    'C': [1, 2, 3, np.nan, 5]
}

df = pd.DataFrame(data)
print("Dữ liệu ban đầu:")
print(df)

# Chuyển các giá trị khuyết về NaN
df.replace(['unknown', 'null', 0], np.nan, inplace=True)

# Điền khuyết bằng mean của từng cột
df_filled = df.fillna(df.mean())

print("\nDữ liệu sau khi điền khuyết bằng mean:")
print(df_filled)
```

### Điền khuyết bằng median

```python
# Tạo DataFrame mẫu với một số giá trị bị khuyết
data = {
    'A': [1, 2, 'null', 4, 5],
    'B': [np.nan, 2, 3, 4, 'unknown'],
    'C': [1, 2, 3, np.nan, 5]
}

df = pd.DataFrame(data)
print("Dữ liệu ban đầu:")
print(df)

# Chuyển các giá trị khuyết về NaN
df.replace(['unknown', 'null', 0], np.nan, inplace=True)

# Điền khuyết bằng mean của từng cột
df_filled = df.fillna(df.median())

print("\nDữ liệu sau khi điền khuyết bằng mean:")
print(df_filled)
```
